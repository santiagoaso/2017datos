```python
dateparseweather = lambda x: pd.datetime.strptime(x, '%m/%d/%Y')
weather = pd.read_csv('./data/weather.csv', parse_dates={'datetime': ['date']}, date_parser=dateparseweather)
weatherparsed = weather
weatherparsed['datetime'] = weatherparsed['datetime'].dt.strftime('%Y%m%d')
weatherparsed[:3]
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')

trips = pd.read_csv('./data/trip.csv', parse_dates={'date': ['start_date']}, date_parser=dateparse)

tripsparsed = trips
tripsparsed['date'] = trips['date'].dt.strftime('%w')
tripsparsed[:3]
```

```python
tpSmall = tripsparsed.loc[:, ('date','subscription_type')].sort_values(by='date')
groupedSubscriber = tpSmall.loc[tpSmall['subscription_type'] == 'Subscriber',('date', 'viajes_subscribers')].groupby('date').agg(np.size)
groupedCustomers = tpSmall.loc[tpSmall['subscription_type'] == 'Customer',('date', 'viajes_customers')].groupby('date').agg(np.size)
print(groupedSubscriber)
print(groupedCustomers)
```

```python
both = pd.concat([groupedSubscriber, groupedCustomers],axis=1)
both
```

```python
graphic1 = both.plot.bar(color=['g','r'])
plt.xlabel('Dia de la semana')
plt.ylabel("Viajes")
plt.title("Cantidad de bicicletas alquiladas por dia de la semana por subscribers y customers")
day_names = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
graphic1.set_xticklabels(day_names,rotation='horizontal')
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')

tripshour = pd.read_csv('./data/trip.csv', parse_dates={'date': ['start_date']}, date_parser=dateparse)

tripsparsedhour = tripshour
tripsparsedhour['date'] = tripshour['date'].dt.hour
tripsparsedhour[:3]
tpSmallhour = tripsparsedhour.loc[:, ('date','subscription_type')].sort_values(by='date')
groupedSubscriberhour = tpSmallhour.loc[tpSmallhour['subscription_type'] == 'Subscriber',('date', 'viajes_subscribers')].groupby('date').agg(np.size)
groupedCustomershour = tpSmallhour.loc[tpSmallhour['subscription_type'] == 'Customer',('date', 'viajes_customers')].groupby('date').agg(np.size)
print(groupedSubscriberhour)
print(groupedCustomershour)
```

```python
bothhour = pd.concat([groupedSubscriberhour, groupedCustomershour],axis=1)
bothhour
```

```python
graphic2 = bothhour.plot.bar(color=['g','r'])
plt.xlabel('Hora del dia')
plt.ylabel("Viajes")
plt.title("Cantidad de bicicletas alquiladas por hora de la semana por subscribers y customers")
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')

tr = pd.read_csv('./data/trip.csv', parse_dates={'date': ['start_date']}, date_parser=dateparse)
tr['date'] = tr['date'].dt.strftime('%Y%m%d')
trSmall = tr.loc[:, ('date','subscription_type')].sort_values(by='date').groupby('date').agg(np.size)
trSmall.head()
```

```python
dateparseweather = lambda x: pd.datetime.strptime(x, '%m/%d/%Y')
wt = pd.read_csv('./data/weather.csv', parse_dates={'datetime': ['date']}, date_parser=dateparseweather)
wt['datetime'] = wt['datetime'].dt.strftime('%Y%m%d')
wtSmall = wt.loc[:, ('datetime','events')].sort_values(by='datetime')
```

```python
wtSmall['events'] = wtSmall['events'].str.lower()
wtSmall['events'].value_counts()
```

```python
wtCount = wtSmall.groupby('datetime').count()
wtCount[:3]
```

```python
#cantidad de eventos por dia separados por aÃ±o
wtCount2013 = wtCount.loc[wtCount.index.str.contains('2013'), :]
wtCount2014 = wtCount.loc[wtCount.index.str.contains('2014'), :]
wtCount2014a = wtCount2014.loc[wtCount2014.index<'20140701', :]
wtCount2014b = wtCount2014.loc[wtCount2014.index>='20140701', :]
wtCount2015 = wtCount.loc[wtCount.index.str.contains('2015'), :]
```
