```python
# magic function para hacer que los graficos de matplotlib se renderizen en el notebook.
%matplotlib inline

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
#Buscamos relacioens entre la falta de servicio y el clima.
weather = pd.read_csv('./data/weather.csv', low_memory=False, parse_dates=['date'])
weather.date = weather.date.dt.date
weather.to_csv("auxWeather.csv")
weather = pd.read_csv('./auxWeather.csv', low_memory=False,)
weather = weather.loc[:,["date" , "max_temperature_f", "mean_temperature_f" , "min_temperature_f", "max_dew_point_f", "mean_dew_point_f", "min_dew_point_f", "max_humidity", "mean_humidity", "min_humidity", "max_sea_level_pressure_inches", "mean_sea_level_pressure_inches", "min_sea_level_pressure_inches", "max_visibility_miles", "mean_visibility_miles", "min_visibility_miles", "max_wind_Speed_mph", "mean_wind_speed_mph", "max_gust_speed_mph", "precipitation_inches", "cloud_cover", "events", "wind_dir_degrees", "zip_code"]]
weather[:10]
```

```python
NotDocks = pd.read_csv('./NotDocks.csv', low_memory=True)
NotDocks['date'] = NotDocks['time']
NotDocks = NotDocks.loc[:,['date','id','zipCode']]
NotDocks[:10]
```

```python
NotBike = pd.read_csv('./NotBike.csv', low_memory=True)
NotBike['date'] = NotBike['time']
NotBike = NotBike.loc[:,['date','id','zipCode']]
NotBike[:10]
```

```python
# Divido por ciudad.
NotBike95113 = NotBike.groupby('zipCode')
NotBike95113 = NotBike95113.get_group(95113)
NotBike95113[:10]
```

```python
#Calculo cuantas veces al dia no hay bisicletas disponibles
DataBike = NotBike95113['date'].value_counts()
df_a = pd.DataFrame(DataBike)
df_a.to_csv("aux95113.csv")     
aux95113 = pd.read_csv('aux95113.csv', low_memory=False)
aux95113['cantidad'] = aux95113['date']
aux95113 = aux95113.loc[:,['cantidad','Unnamed: 0']]
aux95113['date'] = aux95113['Unnamed: 0']
aux95113 = aux95113.loc[:,['cantidad','date']]
aux95113[:10] 
```

```python
#join
aux95113 = pd.merge(NotBike95113, aux95113, on='date', how='left')
aux95113[:10]
```

```python
#divido el clima por ciudad
weather95113 = weather.groupby('zip_code')
weather95113 = weather95113.get_group(95113)
weather95113[:5]
```

```python
#join
salida = pd.merge(weather95113, aux95113, on='date', how='inner')
salida[:5]
```

```python
#Busco relaciones 
#Como es de esperar no encuentro ninguna relacion fuerte.
salida.corr().loc[:,['cantidad']]
```
