```python
#Para que no se vuelva demaciado largo lo separo.
%matplotlib inline

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
status = pd.read_csv('./data/status.csv', low_memory=True)
status[:10]
```

```python
status= status.loc[:,['station_id','bikes_available','docks_available','time']].groupby('station_id')
statusAux = status.get_group(2)
statusAux[:10]
```

```python
tabla = pd.read_csv('./tablIdZip.csv', low_memory=True)
tabla['station_id'] = tabla['id']
tabla[:10]
```

```python
statusAux = pd.merge(statusAux, tabla, on='station_id', how='right')
statusAux[:10]
```

```python
statusAux['time'] = pd.to_datetime(statusAux['time'])
statusAux['year'] = statusAux['time'].map(lambda x:x.year)
statusAux['hour'] = pd.to_datetime(statusAux['time'], format = '%H', errors = 'coerce').dt.hour

statusAux[:10]
```

```python
statusAuxf = statusAux.loc[:,['station_id','id','bikes_available','docks_available','time','zipCode','year','hour']]

statusAuxf[:10]
```

```python
statusAuxf.to_csv("StatusResum2.csv")
```
